```python
import pandas as pd
import numpy as np
import seaborn as sns
import os
path = os.getcwd()
```

```python
path
```

```python
topic1_scores = pd.read_csv(path + '\\topic1_scores.csv')
topic1_all = pd.read_csv(path + '\\topic1_all.csv')
summary = pd.read_csv(path + '\\summary.csv')
```

```python
topic1_scores.head()
topic1_scores.info()
topic1_scores.describe().T
```

### Findings

Out of all the columns, 1.3 Challenge has the lowest mean and highest standard deviation. The mean challenge total is approximately 89.11, with a higher standard deviation of about 30.30 compared to other metrics. The minimum is 0, and the maximum is 100. It seems that the challenge total has a wider variation in scores compared to participation and overall grade.

```python
sns.countplot(topic1_scores['1.3 - Challenge (6)'])
```

```python
import matplotlib.pyplot as plt

num_cols = ['Percent grade', 'Points earned (out of 49)',
       'Participation total (43)', 'Challenge total (6)',
       '1.1 - Participation (9)', '1.2 - Participation (9)',
       '1.3 - Participation (9)', '1.4 - Participation (16)',
       '1.3 - Challenge (6)']

for col in num_cols:
    print(col)
    print('Skew :', round(topic1_scores[col].skew(), 2))
    plt.figure(figsize = (15, 4))
    plt.subplot(1, 2, 1)
    topic1_scores[col].hist(grid=False)
    plt.ylabel('count')
    plt.subplot(1, 2, 2)
    sns.boxplot(x=topic1_scores[col])
    plt.show()
```

```python
plt.figure(figsize=(13,17))
sns.pairplot(data=topic1_scores)
plt.show()
```
